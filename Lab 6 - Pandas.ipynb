## (3) Pandas Basics

#### Pandas Official Documentation: https://pandas.pydata.org/pandas-docs/stable/


```python
import pandas as pd
filename = 'Crimes_-_2001_to_present.csv'
```

```python
# Read the data in to a Pandas DataFrame
crimes_df = pd.read_csv(filename)
```

```python
# Checking the Type of a DataFrame
print('\nThis is the type of the DataFrame')
print(type(crimes_df))
print('--------------------------------------------------')

# Checking the Number of Records in a DataFrame
print('\nThis is the number of records in the DataFrame')
print(len(crimes_df))
print('--------------------------------------------------')

# Print first 'n' records (Similar to "head -n ..." from Shell script)
print('\nThese are the first 5 records in the DataFrame')
print(crimes_df[:5])
print('--------------------------------------------------')

# Print the 'nth' record
print('\nThis is the "10th" record in the DataFrame')
print(crimes_df.iloc[10])
print('--------------------------------------------------')

# Check the DataFrame Index
print('\nThis is the DataFrame Index')
print(crimes_df.index)
print('--------------------------------------------------')

# Check the Columns in a DataFrame
print('\nThis is the list of Columns in the DataFrame')
print(crimes_df.columns)
print('--------------------------------------------------')

# Print the specific columns
#print('\nThese are specific columns in the DataFrame')
#print(crimes_df[["Case Number", "Primary Type", "Description"]])
#print('--------------------------------------------------')
```

```python
# Quick Summary Statistic for a given Column
print('\nCategorical Variable')
print('These are quick summary stats for the Arrest column in the DataFrame')
print(crimes_df['Arrest'].describe())
print('\nNumerical Variable')
print('These are quick summary stats for the Year column in the DataFrame')
print(crimes_df['Year'].describe())
print('\n NOTE: Key Difference in Outputs for a Numerical Variable & a Categorical Variable!')
print('--------------------------------------------------')

# Histogram (Frequency Data!)
print('\nThis is the data for plotting a Histogram for the specific Column in the DataFrame')
print(crimes_df['Arrest'].value_counts())
print('\n NOTE: Items are always sorted by Descending Frequency!')
print('  - For ascending use Reverse Slicing or check Pandas documentation for more options of the value_counts function')
print('--------------------------------------------------')

# Use Head / Tail to get Most Frequent & Least Frequent!
print('\nThese are Most Frequent Year in the DataFrame => Head Function')
print(crimes_df['Year'].value_counts().head(1))
print('\nThese are Least Frequent Year in the DataFrame => Tail Function')
print(crimes_df['Year'].value_counts().tail(1))
print('--------------------------------------------------')
```

```python
# Filtering based on Variable
print('\nThese are records in the DataFrame where Arrest == True')
print(crimes_df[crimes_df['Arrest']])
print('--------------------------------------------------')

# Division of Series A / Series B
```

```python
# Pandas Filters a DataFrame based on the "Index"
#  If the Index of df_1 and df_2 match you can directly use df_1[df_2] !

df_1 = crimes_df[crimes_df['Arrest']]['Primary Type'].value_counts()
df_2 = crimes_df[crimes_df['Arrest']]['Primary Type'].value_counts() >= 5000

print('\n Series 1\n', df_1)
print('\n Series 2\n', df_2)
print('\n Series 3: Filtered\n', df_1[df_2])
```

```python
# Location of Crime with Highest Number of Arrests
arrests = crimes_df[crimes_df['Arrest'] == True]
arrests['Location Description'].value_counts()
```
