```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

crime_file = 'Crimes_-_2001_to_present.csv'
community_file = 'Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012.csv'
```

```python
crimes_df = pd.read_csv(crime_file)
community_df = pd.read_csv(community_file)
#community_df = community_df.dropna(subset=['Community Area Number'])
```

```python
crimes_df['Primary Type'].unique()
#community_df.head(5)
```

```python
community_df = community_df.dropna(subset=['Community Area Number'])
```

```python
theft_counts = crimes_df[crimes_df['Primary Type'] == 'THEFT']['Community Area'].value_counts().reset_index()
#crime_counts = crimes_df['Community Area'].value_counts().reset_index()
```

```python
theft_counts.columns = ['Community Area', 'Crime Count']
#crime_counts.columns = ['Community Area', 'Crime Count']
```

```python
theft_counts.head(3)
#crime_counts.head(3)
```

```python
theft_counts_community = crime_counts.merge(community_df, left_on='Community Area', right_on='Community Area Number', how='right')
#crime_counts_community = crime_counts.merge(community_df, left_on='Community Area', right_on='Community Area Number')
```

```python
theft_counts_community
#theft_counts_community[theft_counts_community['PER CAPITA INCOME '].isnull()]
```

```python
theft_counts_community['Crime Count'].fillna(0, inplace=True)
```

```python
theft_counts_community.plot(kind='scatter', y='Crime Count', x='PER CAPITA INCOME ')
#crime_counts_community.plot(kind='scatter', y='Crime Count', x='PER CAPITA INCOME ')
```
